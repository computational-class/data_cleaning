### DATA WRANGLING

```python
import pandas as pd 
degrees = pd.read_csv('data_salary.csv')
degrees.head()
```

```python
degrees.columns
```

```python
degrees.columns = ['major','Start Med Sal', 'Mid Career Med Sal', 'Start-Mid Career Sal Change', '10%tile Sal','25%tile Sal','75%tile Sal','90%tileSal']
```

```python
degrees.head()
```

```python
def clean(salaries):
    no_dollar = salaries.replace('$','')
    no_comma = no_dollar.replace(',','')
    cleaned  = float(no_comma)
    return(cleaned)
```

```python
degrees.columns
```

```python
salary_columns = ['Start Med Sal','Mid Career Med Sal','10%tile Sal','25%tile Sal','75%tile Sal','90%tileSal']

for i in salary_columns:
    degrees[i] = degrees[i].apply(clean)
```
